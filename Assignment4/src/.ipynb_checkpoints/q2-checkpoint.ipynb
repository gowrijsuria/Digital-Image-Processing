```python
import cv2 
import matplotlib.pyplot as plt 
import numpy as np 
from skimage import measure
from skimage import filters
from skimage.morphology import disk 
import random
import sys 
sys.setrecursionlimit(100000)

def opening(im,disk):
    eroded = cv2.erode(im,disk)
    dil = cv2.dilate(eroded,disk)
    return dil
    
def closing(im,disk):    
    dil = cv2.dilate(im,disk)
    eroded = cv2.erode(dil,disk)
    return eroded
    
def dfs(im,i,j,c,colored):
    if (i<0 or i>=im.shape[0] or j<0 or j>=im.shape[1] or marked[i,j]==1):
        return 0
    
    marked[i,j]=1
    if im[i,j]==1:
        colored[i,j] = c
        dfs(im,i-1,j,c,colored)
        dfs(im,i,j-1,c,colored)
        dfs(im,i+1,j,c,colored)
        dfs(im,i,j+1,c,colored)
    return 0
    
kidney = cv2.imread('../input/kidney.jpg',0)
h = kidney.shape[0]
w = kidney.shape[1]

label_cnt = 0 
label = np.zeros((h,w),dtype='int')

thresh = 130
im_bw = cv2.threshold(kidney, thresh,1, cv2.THRESH_BINARY)[1]
# im_bw = (kidney>130).astype(np.uint8)
print(np.amax(im_bw))


disk6 = disk(6)
disk1 = disk(1)

closed = closing(im_bw,disk1)
opened = opening(closed,disk6)
opened = closing(opened,disk1)

plt.figure(figsize = (15,15))
plt.subplot(1,2,1)
plt.imshow(kidney,'gray')
plt.subplot(1,2,2)
plt.imshow(opened,'gray')

# marked=np.zeros((h,w))
# colored = np.zeros((h,w,3),dtype=np.uint16)
# colors_arr = []

# for i in range(40):
#     new_color = [random.randint(0,255), random.randint(0,255),random.randint(0,255)]
#     colors_arr.append(new_color)
# # print(colors_arr)    
# p=1
# for i in range(h):
#     for j in range(w):
#         if (im_bw[i,j]==1 and marked[i,j]==0):
#             dfs(im,i,j,colors_arr[j%40])
#             p=p+1
            
# plt.figure()            
# # plt.imshow(cv2.cvtColor(colored,cv2.COLOR_BGR2RGB))            
# plt.imshow((colored).astype(np.uint16))
```

```python
marked=np.zeros((h,w))
colored = np.zeros((h,w,3),dtype=np.uint16)
colors_arr = []

for i in range(40):
    new_color = [random.randint(0,255), random.randint(0,255),random.randint(0,255)]
    colors_arr.append(new_color)
# print(colors_arr)    
p=1
for i in range(h):
    for j in range(w):
        if (im_bw[i,j]==1 and marked[i,j]==0):            
            dfs(opened,i,j, colors_arr[p%40],colored)
            p=p+1
#             
plt.figure()            
# plt.imshow(cv2.cvtColor(colored,cv2.COLOR_BGR2RGB))            
plt.imshow(colored)
```
