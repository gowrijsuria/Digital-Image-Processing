```python
import numpy as np 
import cv2 
import matplotlib.pyplot as plt
from scipy import signal 
from math import exp
 
im = cv2.imread('../input/lena.jpg',0)
h,w = im.shape
H_spatial = np.array([[0,1,0],[1,2,1],[0,1,0]])
eq_filt = np.fft.fftshift(np.fft.fft2(H_spatial))
print(eq_filt)

pad_h = np.zeros((3+h-1,3+w-1))
pad_h[0:3,0:3] = H_spatial
pad_im = np.zeros((h+2,w+2))
pad_im[0:h,0:w] = im

# X = np.arange(h)
# Y = np.arange(w)
# (r,c) = np.meshgrid(X,Y)
# j = np.complex(0,1)
# eq_filt = np.zeros((h,w))
# eq_filt = 2+ exp(np.pi*2*j*r/X) + exp(np.pi*-2*j*r/X) + exp(np.pi*-2*j*c/Y) + exp(np.pi*2*j*c/Y)
# p = np.multiply(eq_filt,np.fft.ftt2(pad_im))
# im_formula = np.fft.ifft2(p)

# pad_h[h//2-2:h//2+1,w//2-2:w//2+1] = H_spatial


mult = np.multiply(np.fft.fft2(pad_h),np.fft.fft2(pad_im))
im_new = (np.fft.ifft2(mult)).real
im_in = signal.convolve2d(im,H_spatial)

plt.figure(figsize = (15,15))
plt.subplot(141)
plt.imshow(im,'gray')
plt.title('original')
plt.subplot(142)
plt.imshow(im_new,'gray')
plt.title('multiplication of ffts')
plt.subplot(143)
plt.imshow(im_in,'gray')
plt.title('convolution thm.')
plt.subplot(144)
plt.imshow(20*np.log(1+abs(eq_filt)),'gray')
plt.title('Equivalent of spatial filter (magnitude)')
plt.show()
```
