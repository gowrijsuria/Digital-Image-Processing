```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def fft(img):
    f = np.fft.fft2(img)
    fshift = np.fft.fftshift(f)
    magnitude_spec = 20*np.log(np.abs(fshift))
    plt.figure(figsize=(10,10))
    plt.subplot(121),plt.imshow(img,'gray')
    plt.title('Input Image'), plt.xticks([]), plt.yticks([])
    plt.subplot(122),plt.imshow(magnitude_spec,'gray')
    plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
    plt.show()

rect = cv2.imread('../input/rectangle.jpg',0)
fft(rect)

h = rect.shape[0]
w = rect.shape[1]

M = cv2.getRotationMatrix2D((w/2,h/2),30,1.0)
rotated = cv2.warpAffine(rect, M, (w, h))
fft(rotated)
```

```python
# translated = np.zeros(h,w)
translation_matrix = np.float32([ [1,0,60], [0,1,0] ])
img_translation = cv2.warpAffine(rect, translation_matrix, (w,h))

fft(rect)
fft(img_translation)
```
