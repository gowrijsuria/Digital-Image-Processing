```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt 

def butterworth(img):
    m = img.shape[0]
    n = img.shape[1]    
    d0 = 10
    order = 1
    filt = np.zeros((m,n))
    f = np.fft.fft2(img)
    fshift = np.fft.fftshift(f)
        
    for i in range(m):
        for j in range(n):
            d = np.sqrt((i - m/2)**2+(j - n/2)**2)
            filt[i,j] = 1/(1 + (d/d0)**(2*order))
 
#     low_fshift = np.fft.fftshift(filt)
    filtered = np.multiply(filt,fshift)
    filtered_ = np.fft.ifft2(np.fft.ifftshift(filtered)).astype('float64')
    plt.figure(figsize=(10,10))
    plt.subplot(121),plt.imshow(img,'gray')
    plt.title('original')
    plt.subplot(122),plt.imshow(filtered_,'gray')
    plt.title('butterworth filter')
    plt.show()
    
def gaussian_filt(img,d0):
    m = img.shape[0]
    n = img.shape[1]    
    filt = np.zeros((m,n))
    f = np.fft.fft2(img)
    fshift = np.fft.fftshift(f)
        
    for i in range(m):
        for j in range(n):
            d = np.sqrt((i - m/2)**2+(j - n/2)**2)
            filt[i,j] = np.exp(-(d**2)/(2*(d0**2)))
 
#     low_fshift = np.fft.fftshift(filt)
    filtered = np.multiply(filt,fshift)
    filtered_ = np.fft.ifft2(np.fft.ifftshift(filtered)).astype('float64')
    plt.figure(figsize=(10,10))
    plt.subplot(121),plt.imshow(img,'gray')
    plt.title('original')
    plt.subplot(122),plt.imshow(filtered_,'gray')
    plt.title('gaussian filtered with sigma {}'.format(d0))
    plt.show()
    return filtered_
            
    
def low_pass(img):
    m = img.shape[0]
    n = img.shape[1]
    low_filt = np.zeros((m,n))
    filtered = np.zeros((m,n))
    cutoff = 350
    f = np.fft.fft2(img)
    fshift = np.fft.fftshift(f)
    
    for i in range(m):
        for j in range(n):
            d = np.sqrt((i - m/2)**2+(j - n/2)**2)
            if d >= cutoff:
                low_filt[i,j] = 1
            else:
                low_filt[i,j] = 0
    
    low_fshift = np.fft.fftshift(low_filt).astype('float64')
    
    filtered = np.multiply(low_fshift,fshift)
    
    filtered_ = np.fft.ifft2(np.fft.ifftshift(filtered)).astype('float64')
    magnitude_spec = 20*np.log(np.abs(low_fshift))
    
    plt.figure(figsize=(10,10))
#     cv2.normalize(filtered_,filtered_, 0, 255, cv2.NORM_MINMAX)
#     filtered_ = filtered_.astype('uint8')
    plt.subplot(121),plt.imshow(img,'gray')
    plt.title('original')
    plt.subplot(122),plt.imshow(filtered_,'gray')
    plt.title('ideal low pass filter')
    plt.show()
    
lena = cv2.imread('../input/lena.jpg',0)
# cutoff_freq = fft(lena)
low_pass(lena)
#
butterworth(lena)


im1 = gaussian_filt(lena,30)
im2 = gaussian_filt(lena,15)
diff = abs(im1 - im2)
plt.imshow(diff,'gray')
plt.title('diff of images')
```
