```python
import cv2 
import matplotlib.pyplot as plt 
import numpy as np 

laplacian = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
blur = cv2.imread('blur.jpg',0)
blur_l = cv2.filter2D(blur,cv2.CV_8U,laplacian)
output = blur - blur_l

gauss = cv2.GaussianBlur(blur, (7,7), 0)
unsharp_image = cv2.addWeighted(blur, 2, gauss, -1, 0)

plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
plt.imshow(blur_l,'gray')
plt.subplot(1,2,2)
plt.imshow(unsharp_image,'gray')
plt.show()
```
