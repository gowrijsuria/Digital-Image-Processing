```python
import cv2 
import matplotlib.pyplot as plt 
import numpy as np 
from math import exp
    
def calc_wt(im_part_b,im_part_g,im_part_r,vard,varr,filt_size):    
    w = np.ones((filt_size,filt_size,3))
    a = np.arange(-(filt_size//2),(filt_size//2)+1)
    repeat_wind_row = np.tile(a,(filt_size,1))
    repeat_wind_col = np.transpose(repeat_wind_row)
    
    cent_x=filt_size//2
    cent_y=filt_size//2
    
    range_f_b = np.zeros((filt_size,filt_size))
    range_f_g = np.zeros((filt_size,filt_size))
    range_f_r = np.zeros((filt_size,filt_size))
    domain_f_b = np.zeros((filt_size,filt_size))
    domain_f_g = np.zeros((filt_size,filt_size))
    domain_f_r = np.zeros((filt_size,filt_size))
            
    s_b = np.full((filt_size,filt_size),im_part_b[cent_x,cent_y])
    s_g = np.full((filt_size,filt_size),im_part_g[cent_x,cent_y])
    s_r = np.full((filt_size,filt_size),im_part_r[cent_x,cent_y])
    one_mat = np.ones((filt_size,filt_size))
    range_f_b[0:filt_size,0:filt_size] = np.exp(-((one_mat[0:filt_size,0:filt_size] - (im_part_b[0:filt_size,0:filt_size] - s_b[0:filt_size,0:filt_size]))**2)/(2*(varr**2)))
    range_f_g[0:filt_size,0:filt_size] = np.exp(-((one_mat[0:filt_size,0:filt_size] - (im_part_g[0:filt_size,0:filt_size] - s_g[0:filt_size,0:filt_size]))**2)/(2*(varr**2)))
    range_f_r[0:filt_size,0:filt_size] = np.exp(-((one_mat[0:filt_size,0:filt_size] - (im_part_r[0:filt_size,0:filt_size] - s_r[0:filt_size,0:filt_size]))**2)/(2*(varr**2)))
    domain_f_b[0:filt_size,0:filt_size] = np.exp(-((repeat_wind_col[0:filt_size,0:filt_size])**2 + (repeat_wind_row[0:filt_size,0:filt_size])**2)/(2*(vard**2)))
    domain_f_g[0:filt_size,0:filt_size] = np.exp(-((repeat_wind_col[0:filt_size,0:filt_size])**2 + (repeat_wind_row[0:filt_size,0:filt_size])**2)/(2*(vard**2)))
    domain_f_r[0:filt_size,0:filt_size] = np.exp(-((repeat_wind_col[0:filt_size,0:filt_size])**2 + (repeat_wind_row[0:filt_size,0:filt_size])**2)/(2*(vard**2)))
            
    w[:,:,0] = np.multiply(domain_f_b,range_f_b)    
    w[:,:,1] = np.multiply(domain_f_g,range_f_g)    
    w[:,:,2] = np.multiply(domain_f_r,range_f_r)    
    return w
    
def cross_bil(flash,no_flash,vard,varr,filt_size):
    h = flash.shape[0]
    w = flash.shape[1]
    output = np.zeros((h,w,3))
    pad = filt_size//2
    pad_im1 = np.zeros((h+2*pad,w+2*pad,3))
    pad_im1[pad:h+pad,pad:w+pad] = no_flash
#     pad_im1[pad:h+pad,pad:w+pad,0] = flash[:,:,0]
#     pad_im1[pad:h+pad,pad:w+pad,1] = flash[:,:,1]
#     pad_im1[pad:h+pad,pad:w+pad,2] = flash[:,:,2]
    pad_im2 = np.zeros((h+2*pad,w+2*pad,3))
    pad_im2[pad:h+pad,pad:w+pad] = flash
#     pad_im2[pad:h+pad,pad:w+pad,0] = no_flash[:,:,0]
#     pad_im2[pad:h+pad,pad:w+pad,1] = no_flash[:,:,1]
#     pad_im2[pad:h+pad,pad:w+pad,2] = no_flash[:,:,2]
    new_h = pad_im1.shape[0]
    new_w = pad_im1.shape[1]
    print(new_h-filt_size,new_w-filt_size)

    for i in range(new_h-filt_size):
        for j in range(new_w-filt_size):
#             print(i,j)
#             print('hi')
            im_part1_b=pad_im1[i:i+filt_size,j:j+filt_size,0]
            im_part1_g=pad_im1[i:i+filt_size,j:j+filt_size,1]
            im_part1_r=pad_im1[i:i+filt_size,j:j+filt_size,2]
            im_part2_b=pad_im2[i:i+filt_size,j:j+filt_size,0]
            im_part2_g=pad_im2[i:i+filt_size,j:j+filt_size,1]
            im_part2_r=pad_im2[i:i+filt_size,j:j+filt_size,2]
        
            new_weight = np.zeros((filt_size,filt_size,3))
            new_weight[:,:,:] = calc_wt(im_part2_b,im_part2_g,im_part2_r,vard,varr,filt_size)        
            num_b = np.sum(np.multiply(im_part1_b,new_weight[:,:,0])) 
            num_g = np.sum(np.multiply(im_part1_g,new_weight[:,:,1])) 
            num_r = np.sum(np.multiply(im_part1_r,new_weight[:,:,2])) 
            den_b = np.sum(new_weight[:,:,0])
            den_g = np.sum(new_weight[:,:,1])
            den_r = np.sum(new_weight[:,:,2])
            output[i,j,0] = num_b/den_b
            output[i,j,1] = num_g/den_g
            output[i,j,2] = num_r/den_r
    return output
    
    
    
    
def bilateral_l(im,vard,varr,filt_size):
    h = im.shape[0]
    w = im.shape[1]
#     print(h,w)
#     img = np.pad(img,(7//2,7//2),mode='constant')
    output = np.zeros((h,w))
    pad = filt_size//2
    pad_im = np.zeros((h+2*pad,w+2*pad))
    pad_im[pad:h+pad,pad:w+pad] = im
    new_h = pad_im.shape[0]
    new_w = pad_im.shape[1]
    
    for i in range(new_h-filt_size):
        for j in range(new_w-filt_size):
#             print(i,j)

            im_part=pad_im[i:i+filt_size,j:j+filt_size]
            cent_x=filt_size//2
            cent_y=filt_size//2
#             for k in range():
#                 for l in range():
            w = np.ones((filt_size,filt_size))
            a = np.arange(-(filt_size//2),(filt_size//2)+1)
            repeat_wind_row = np.tile(a,(filt_size,1))
            repeat_wind_col = np.transpose(repeat_wind_row)
            
            range_f = np.zeros((filt_size,filt_size))
            domain_f = np.zeros((filt_size,filt_size))
            
            s = np.full((filt_size,filt_size),im_part[cent_x,cent_y])
            one_mat = np.ones((filt_size,filt_size))
            range_f[0:filt_size,0:filt_size] = np.exp(-((one_mat[0:filt_size,0:filt_size] - (im_part[0:filt_size,0:filt_size] - s[0:filt_size,0:filt_size]))**2)/(2*(varr**2)))
            domain_f[0:filt_size,0:filt_size] = np.exp(-((repeat_wind_col[0:filt_size,0:filt_size])**2 + (repeat_wind_row[0:filt_size,0:filt_size])**2)/(2*(vard**2)))
            
            
            w = np.multiply(domain_f,range_f)    
            num = np.sum(np.multiply(im_part,w)) 
            den = np.sum(w)
            output[i,j] = num/den
            
    return output





im = cv2.imread('../input/cubes.png',0)
flash = cv2.imread('../input/pots_flash.jpg')
no_flash = cv2.imread('../input/pots_no_flash.jpg')

cross_b = cross_bil(flash,no_flash,105,10,7)
# cv2.normalize(cross_b[:,:,0],cross_b[:,:,0], 0, 255, cv2.NORM_MINMAX)
# cross_b[:,:,0] = cross_b[:,:,0].astype('uint8')
# cv2.normalize(cross_b[:,:,1],cross_b[:,:,1], 0, 255, cv2.NORM_MINMAX)
# cross_b[:,:,1] = cross_b[:,:,1].astype('uint8')
cv2.normalize(cross_b,cross_b, 0, 255, cv2.NORM_MINMAX)
cross_b = cross_b.astype('uint8')


cross_b_rgb = cv2.cvtColor(cross_b,cv2.COLOR_BGR2RGB)
no_flash = cv2.cvtColor(no_flash,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15,15))
plt.subplot(1,3,1)
plt.imshow(flash)
plt.subplot(1,3,2)
plt.imshow(no_flash)
plt.subplot(1,3,3)
plt.imshow(cross_b_rgb)
plt.show()

output = bilateral_l(im,20,12,7)
plt.figure(figsize=(15,15))
plt.subplot(1,2,1)
plt.imshow(im,'gray')
plt.subplot(1,2,2)
plt.imshow(output,'gray')
plt.show()
```
