```python
barbara = cv2.imread('../input/barbara.jpg',0)

# def conv(im,filt):
#     h = im.shape[0]
#     w = im.shape[1]
#     size = filt.shape[0]
#     if(size%2 != 0):
#         pad = size//2
#         pad_im = np.pad(im,(pad,pad),mode='constant')
#         new_h = pad_im.shape[0]
#         new_w = pad_im.shape[1]
#     else:
#         pad_im = im
#         new_h = pad_im.shape[0]
#         new_w = pad_im.shape[1]
        
#     kernel = np.flipud(np.fliplr(filt))
#     output_im = np.zeros((h,w))
#     filt_h = filt.shape[0]
#     filt_w = filt.shape[1]
#     start_w = filt_w//2
#     start_h = filt_h//2

#     for j in range(start_h,new_h - start_h):
#         for i in range(start_w,new_w - start_w):
#             print(im[i-start_w:i+start_w,j-start_h:j+start_h].shape)
#             output_im[i,j] = np.sum(np.multiply(im[i-start_w:i+start_w,j-start_h:j+start_h],kernel))
#             print(i,j)
#     print(output_im)

    
def conv3(im,filt):
    
    h = im.shape[0]
    w = im.shape[1]
    pad = 1
    pad_im = np.pad(im,(pad,pad),mode='constant')
    new_h = pad_im.shape[0]
    new_w = pad_im.shape[1]
    output = np.zeros((h,w))
    kernel = np.flipud(np.fliplr(filt))

    for i in range(1,h-1):
        for j in range(1,w-1):
            output[j][i] = np.sum(im[j-1:j+2,i-1:i+2]*kernel)
    
    return output   


def sobel(im):
    mask_x = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
    mask_y = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
    output_x = conv3(im,mask_x)
    output_y = conv3(im,mask_y)
    output = np.hypot(output_x,output_y)
    return output

def roberts(im):
    mask_x = np.array([[0,0,0],[0,0,1],[0,-1,0]])
    mask_y = np.array([[0,0,0],[0,1,0],[0,0,-1]])
    output_x = conv3(im,mask_x)
    output_y = conv3(im,mask_y)
    output = np.hypot(output_x,output_y)
    return output

def prewitt(im):
    mask_x = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
    mask_y = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
    output_x = conv3(im,mask_x)
    output_y = conv3(im,mask_y)
    output = np.hypot(output_x,output_y)
    
    return output

def laplacian(im):
    mask = np.array([[1,1,1],[1,-8,1],[1,1,1]])
    output = conv3(im,mask)
    return output

b_canny = cv2.Canny(barbara,120,10)

b_robert = roberts(barbara)
b_sobel = sobel(barbara)
b_prewitt = prewitt(barbara)
b_laplacian = laplacian(barbara)


plt.figure(figsize = (15,15))
plt.subplot(3,2,1)
plt.title('original')
plt.imshow(barbara,'gray')
plt.subplot(3,2,2)
plt.title('canny')
plt.imshow(b_canny,'gray')
plt.subplot(3,2,3)
plt.title('robert')
plt.imshow(b_robert,'gray')
plt.subplot(3,2,4)
plt.title('sobel')
plt.imshow(b_sobel,'gray')
plt.subplot(3,2,5)
plt.title('prewitt')
plt.imshow(b_prewitt,'gray')
plt.subplot(3,2,6)
plt.title('laplacian')
plt.imshow(b_laplacian,'gray')
plt.show()
```

```python
sigma = 5
width,height = barbara.shape
noise = np.random.randn(width,height)*sigma
add_noise = np.uint8(barbara + noise)

b_canny_n = cv2.Canny(add_noise,20,100)
b_robert_n = roberts(add_noise)
b_sobel_n = sobel(add_noise)
b_prewitt_n = prewitt(add_noise)
# b_laplacian_n = cv2.Laplacian(add_noise,cv2.CV_64F)
b_laplacian_n = laplacian(add_noise)

plt.figure(figsize = (15,15))
plt.subplot(3,2,1)
plt.title('original with gaussian noise')
plt.imshow(add_noise,'gray')
plt.subplot(3,2,2)
plt.title('canny after adding noise')
plt.imshow(b_canny_n,'gray')
plt.subplot(3,2,3)
plt.title('robert after adding noise')
plt.imshow(b_robert_n,'gray')
plt.subplot(3,2,4)
plt.title('sobel after adding noise')
plt.imshow(b_sobel_n,'gray')
plt.subplot(3,2,5)
plt.title('prewitt after adding noise')
plt.imshow(b_prewitt_n,'gray')
plt.subplot(3,2,6)
plt.title('laplacian after adding noise')
plt.imshow(b_laplacian_n,'gray')
plt.show()
```
